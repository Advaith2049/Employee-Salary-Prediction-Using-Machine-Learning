```python
import pandas as pd
import pickle
from xgboost import XGBRegressor
```

```python
emp = pd.read_csv('cleaned_data.csv',index_col =[0])
emp
```

```python
target = 'salary'
encode = ['jobType','degree','major','industry']
```

```python
for col in encode:
    dummy = pd.get_dummies(emp[col], prefix=col)
    emp = pd.concat([emp,dummy], axis=1)
    del emp[col]
```

```python
emp
```

```python
X = emp.drop('salary',axis = 1)
y = emp['salary']
```

```python
clf = XGBRegressor()
```

```python
clf.fit(X,y)
```

```python
pickle.dump(clf, open('emp_clf.pkl', 'wb'))
```
